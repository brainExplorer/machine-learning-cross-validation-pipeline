```python
# -*- coding: utf-8 -*-
"""
Preventing Data Leakage With Cross-Validation And Pipeline
1- Apply K-Fold CV logic manually and automaticaly
2- To integrate the scaling step with CV using pipeline and prevent data leakage.
"""
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import KFold, train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Seed for repeatable random transactions
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

```python
# --------------------
# Upload Data and Apply as a Training/Test
# --------------------
data = load_breast_cancer()
X = data.data
y = data.target

# We allocate class distribution by maintaining class distribution (90% (90% training, 10% test).
# The test set IS SETTING for final evaluation and NEVER used during CV.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=RANDOM_STATE)

print(f"Training test size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
#print(f"Training set distribution: {np.unique(y_train, return_counts=True)}")
#print(f"Test set distribution: {np.unique(y_test, return_counts=True)}")
print(f"Training set distribution: {np.bincount(y_train)}")
print(f"Test set distribution: {np.bincount(y_test)}")
```

```python
# ------------------------------------------
# K-Fold Cross-Validation (Manual Applying)
# ------------------------------------------

print("K-Fold Cross-Validation (Manual Applying) CV (5 fold)")
kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
model = LogisticRegression(max_iter=1000, random_state=RANDOM_STATE)

fold_scores = []
for fold, (train_idx, val_idx) in enumerate(kf.split(X_train)):
    X_train_fold, X_val_fold = X_train[train_idx], X_train[val_idx]
    y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]

    model.fit(X_train_fold, y_train_fold)
    y_pred = model.predict(X_val_fold)
    score = accuracy_score(y_val_fold, y_pred)
    fold_scores.append(score)
    print(f"Fold {fold + 1} Accuracy: {score:.4f}")

print(f"Mean Accuracy: {np.mean(fold_scores):.4f} (+/- {np.std(fold_scores):.4f})")
```

```python
# cross_val_score with automatic KFold
print("cross_val_score with automatic KFold")
scores = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")
print(f"Mean Accuracy: {np.mean(scores):.4f} (+/- {np.std(scores):.4f})")
```

```python
#-----------------------------------------------------------------------------------------
#Fourth The Importance of Stratified K-Fold (Class Unbalance Status)
#--------------------------------------------------------------------------------------
# This dataset is not too unstable, but let's see the behavior of StratifiedKFold.
# Let's examine the class rates on the floors with KFold first.

print("KFold and StratifiedKFold")
kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

print("\nPositive (1) rates on the floors with KFold")
for fold, (_, val_idx) in enumerate(kf.split(X_train)):
  y_val = y_train[val_idx]
  positive_rate = np.mean(y_val)
  print(f"Fold {fold + 1}: Positive Rate = {positive_rate:.4f}")

print("\nPositive (1) rates on the floors with StratifiedKFold")
for fold, (_, val_idx) in enumerate(skf.split(X_train, y_train)):
  y_val = y_train[val_idx]
  positive_rate = np.mean(y_val)
  print(f"Fold {fold + 1}: Positive Rate = {positive_rate:.4f}")

# StratifiedKFold keeps the class distribution of each floor close to the entire training set.
# This makes model evaluation more reliable, especially in unstable data.
```

```python
#------------------------------------------------------------------------------
# Data Leakage Danger
#------------------------------------------------------------------------------
# WRONG ROAD: First scale the whole training set and then apply CV.

print("WRONG ROAD: First scale the whole training set and then apply CV.")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

scores_leakly = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring="accuracy")
print(f"Mean Accuracy: {np.mean(scores):.4f} (+/- {np.std(scores_leakly):.4f})")
print(f"Average accuracy (without leakage): scores_leaky.mean(:4f' (+/- scores_leaky.std(.4f')")
print("NOT: These results are NOT REAL because each fold's verification set leaked information during scaling.")

#------------------------------------------------------------------
# CORRECT APPLICATION . integrating scaling into CV with Pipeline
#------------------------------------------------------------------

print("CORRECT APPLICATION . integrating scaling into CV with Pipeline")
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000, random_state=RANDOM_STATE))
])

# We give Pipeline directly to cross_val_score.
scores_pipeline = cross_val_score(pipeline, X_train, y_train, cv=5, scoring="accuracy")
print(f"Mean Accuracy: {np.mean(scores_pipeline):.4f} (+/- {np.std(scores_pipeline):.4f})")

```

```python
# ------------------------------------------------
# Final Model Training and Evaluation in Test Set
# ------------------------------------------------
print("Final Model Training and Evaluation in Test Set")
pipeline.fit(X_train, y_train)

# We estimate on the test set (test set was not included in the CV in any way).
y_pred = pipeline.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {test_accuracy:.4f}")
print ("This result is the closest to the model's real-life performance forecast.")
```
